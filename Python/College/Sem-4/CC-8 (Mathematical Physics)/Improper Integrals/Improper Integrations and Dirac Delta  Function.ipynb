#### Modules to Import

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

plt.style.use('dark_background')
np.seterr(divide = 'ignore')
```

$$ \int_0 ^ 1 \ln(x) dx $$

```python
from math import log

I = lambda x: log(x)
sol = quad(I,0,1)
print("The Value of the improper integral is: %0.3f "%sol[0])

X = np.linspace(0.1, 5, 1000, retstep = False)
Y = np.log(X)

plt.plot(X, Y, label = r'$ln(x)$')
plt.axhline(linestyle = '--')
plt.axvline(linestyle = '--')
plt.legend(loc= 'best')
plt.fill_between(X,Y, label = 'Area')
plt.show()
```

## *Dirac's Delta Function*
### $$\int_{-\infty} ^{\infty} f(x)\delta(x-a) = f(a)$$
### $$\int_{-\infty} ^{\infty}\delta(x-a) = 1$$
#### $f(x) = x^2$

```python
a = 3
n = 10
x0 = 6

X = np.linspace(a-x0,a+x0,1000,retstep = False)
Y = np.zeros_like(X)

def f(x):   # The arbitary function
    return x**2 
```

$$\delta{(x-a)} = \lim_{n \to \infty} \sqrt{\frac{n}{\pi}} e ^{-n(x-a)^2}$$

```python
from math import sqrt,exp

def gaussian(x,n):
    return sqrt(n/np.pi)*exp(-n*(x-a)**2)

for i in range(len(X)):
    Y[i] = gaussian(X[i],n)

def g(x):
    return f(x)*gaussian(x,n) 

X1 = np.linspace(-100,100,1000)
Z = np.zeros_like(X1)
for i in range(len(X1)):
    Z[i] = g(X1[i])


val1 = quad(g,-np.infty, np.infty)
val2 = quad(gaussian,-np.infty, np.infty, args = n,)

print("The Value of the first integration f(x)d(x-a) is: %.2f "%val1[0],", where as the Actual Value of f(a) is: ",f(a))
print("\nThe Value of the second integration f(x)  is: %.2f"%val2[0])


plt.subplot(121)
plt.title(r"$\delta(x-a)$")
plt.plot(X,Y)
plt.show()  
```

$$\delta{(x-a)} = \lim_{n \to \infty} = \frac{n}{\pi} \frac{1}{1+n^2(x-a)^2}$$

```python
def delta(x,n):
    return (n/np.pi)*(1/(1+(n*(x-a))**2))

for i in range(len(X)):
    Y[i] = delta(X[i],n)

def g(x):
    return f(x)*delta(x,n) 

X1 = np.linspace(-100,100,1000)
Z = np.zeros_like(X1)
for i in range(len(X1)):
    Z[i] = g(X1[i])

val1 = quad(g,-np.infty, np.infty)
val2 = quad(delta,-np.infty, np.infty, args = n,)

print("The Value of the first integration f(x)d(x-a) is: %.2f "%val1[0],", where as the Actual Value of f(a) is: ",f(a))
print("\nThe Value of the second integration f(x)  is: %.2f"%val2[0])


plt.subplot(121)
plt.title(r"$\delta(x-a)$")
plt.xticks([a])
plt.plot(X,Y)
plt.show()  
```

$$\delta{(x-a)} = \lim_{n \to \infty} \frac{sin\{n(x-a)\}}{\pi(x-a)} $$

```python
from math import sin,pi

def delta(x,n):
    return sin(n*(x-a))/(pi*(x-a))

for i in range(len(X)):
    Y[i] = delta(X[i],n)

def g(x):
    return f(x)*delta(x,n) 

X1 = np.linspace(-100,100,1000)
Z = np.zeros_like(X1)
for i in range(len(X1)):
    Z[i] = g(X1[i])

try:
    val1 = quad(g,-np.infty, np.infty)
except: ZeroDivisionError

try:
    val2 = quad(delta,-np.infty, np.infty, args = n,)
except: ZeroDivisionError

print("The Value of the first integration f(x)d(x-a) is: %.2f "%val1[0],", where as the Actual Value of f(a) is: ",f(a))
print("\nThe Value of the second integration f(x)  is: %.2f"%val2[0])


plt.subplot(121)
plt.title(r"$\delta(x-a)$")
plt.plot(X,Y)
plt.show()  
```

$$\delta{(x-a)} = \lim_{\epsilon \to 0} \frac{1}{2\epsilon \cosh^2\left({\frac{(x-a)}{\epsilon}}\right)}$$

```python
from math import cosh,pow

eps = 0.2

def delta(x,eps):
    return 1/(2*eps*(cosh((x-a)/eps))*eps*(cosh((x-a)/eps)))

for i in range(len(X)):
    Y[i] = delta(X[i],eps)

def g(x):
    return f(x)*delta(x,eps) 

X1 = np.linspace(-100,100,1000)
Z = np.zeros_like(X1)
for i in range(len(X1)):
    Z[i] = g(X1[i])

try:
    val1 = quad(g,-np.infty, np.infty)
except: OverflowError
try:
    val2 = quad(delta,-np.infty, np.infty, args = eps,)
except: OverflowError

print("The Value of the first integration f(x)d(x-a) is: %.2f "%val1[0],", where as the Actual Value of f(a) is: ",f(a))
print("\nThe Value of the second integration f(x)  is: %.2f"%val2[0])

plt.subplot(121)
plt.title(r"$\delta(x-a)$")
plt.plot(X,Y)
plt.show()  
```
