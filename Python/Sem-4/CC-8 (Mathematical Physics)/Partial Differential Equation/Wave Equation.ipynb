### Wave Equation for 1-D $$\frac{\partial^2 U}{\partial x^2} = \frac{1}{c^2}\frac{\partial^2 U}{\partial t^2}$$

Modules to Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

Question 1:
\begin{align*}
\text{Boundary Condition: } &U(0,t) = 0 \text{ and }U(L,t) = 0\\ 
\text{Initial condition: } &U(x,0) = e^{-k(x-x_0)^2}
\end{align*}

Initialization

```python
length = 1  # Length of the String
n = 1001  # number of datapoints on the string
X, dx = np.linspace(0, 1, n, retstep=True)
c = 1  # Velocity of the wave

K, x0 = 1000, 0.5
time = 1  # duration
m = 4001  # time steps
T, dt = np.linspace(0, time, m, retstep=True)

r = c*dt/dx  # Stability factor
U = np.zeros((m, n))
```

Implementation of the boundary and initial conditions

```python
BoundaryConditions = [0, 0]

def InitialCondition(x: list):
    return np.exp(-K*((x-x0))**2)

def g(x):
    return 0

U[0, :] = InitialCondition(X)
U[:, -1] = BoundaryConditions[-1]
U[:, 0] = BoundaryConditions[0]
```

Implementation of Finite Difference Method

```python
for i in range(1, n-1):  # Calculate the displacement of the string at t=1
    U[1, i] = (r**2)*(U[0, i+1]+U[0, i-1]) + 2 * \
        (1-r**2)*U[0, i] - (U[0, i]-g(U[1])*dt)

for j in range(1, m-1):  # Calculate the displacement of the string at t>1
    for i in range(1, n-1):
        U[j+1, i] = (r**2)*(U[j, i+1]+U[j, i-1]) + 2*(1-r**2)*U[j, i]-U[j-1, i]
```

Plotting the displacement of the string at various time

```python
mark = ['o', '*', '.', '<', 'P', 's', '>', 'x', '^', 'D']
for k in range(0, m, 500):
    plt.plot(X, U[k, :], label='t = %0.2f' %
             T[k], marker=mark[int(k/500)], markevery=60)
plt.title('Wave Propagation')
plt.text(0.65, 0.5, r'$\rightarrow$', fontsize=25)
plt.text(0.3, -0.5, r'$\rightarrow$', fontsize=25)
plt.text(0.65, -0.5, r'$\leftarrow$', fontsize=25)
plt.text(0.3, 0.5, r'$\leftarrow$', fontsize=25)
plt.axhline(linestyle='--')
plt.xlabel('Position')
plt.ylabel('Displacement')
plt.legend(loc='best')
#plt.savefig('wave1.png')
plt.show()
```
