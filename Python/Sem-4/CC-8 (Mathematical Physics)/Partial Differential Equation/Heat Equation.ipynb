### Fourier Diffusion Equation for 1-D $$\frac{\partial U}{\partial t} = \alpha\frac{\partial^2 U}{\partial x^2}$$

Modules to Import

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Question 1:- Heat Point is at a side $(x=0)$ of the rod

Creating The grid for the temperature profile at various time at various points

```python
length = 20      #Length of the String
dx = 1/3         # distance between two points on the rod
X = np.arange(0,length,dx)
n = len(X)       #number of positions

time = 100       #duration
dt = 1/36        #smallest time step
T = np.arange(0,time,dt)
m = len(T)          #number of time steps

alpha = 0.25 

U = np.zeros((m,n))     # Creating the grid
```

Implementation of the initial and boundary conditions 

```python

BoundaryCondition=[100,0]
InitialCondition = 0

U[0,:] = InitialCondition 
U[:,0] = BoundaryCondition[0]
U[:,-1] = BoundaryCondition[1]
```

Implementation of the finite difference method

```python
for j in range(0,m-1):        #time axis(axis = 0)
    for i in range(1,n-1):  #position axis(axis = 1)
       U[j+1,i] = alpha*(U[j,i+1] + U[j,i-1]) + (1-2*alpha)*U[j,i]
```

```python
mark=['o','D','P','^','s','>','<','+',',','.','*']
plt.title('Solution of 1D Fourier heat equation when the heatpoint is at x=0')
plt.xlabel('Position along the rod')
plt.ylabel(r'Temperature in [$\degree C$]')
for k in range(0, m-1,600):
    plt.plot(X,U[k,:],label = 't = %d'%T[k],marker = mark[int(k/600)],markevery = 3)
plt.legend(loc='best')
#plt.savefig('heat1.png')
plt.show()
```

#### Question 2:- Heat Point is at the middle of the rod $(x=0.5)$

Creating The grid for the temperature profile at various time at various points

```python
Length = 1  # Length of the Rod
Time = 100  # Total time
dx = 0.05  # distance between two points on the rod
dt = 1  # smallest time step
alpha = 0.25  # thermal conductivity

# Creating The grid for the temperature profile at various time at various points
X = np.arange(0, Length+dx, dx)
T = np.arange(0, Time+dt, dt)

# finding the midpoint of the rod
p = round(len(X)/2)
```

Implementation of Boundary and Initial Conditions

```python
BoundaryCondition = [10, 0]
InitialCondition = 0

m = len(X)
n = len(T)
U = np.zeros((n, m))  # Creating the grid

# implementing the initial and boundary conditions
U[-1, :] = InitialCondition
U[:, p] = BoundaryCondition[0]
U[:, -1] = BoundaryCondition[1]
```

Implementation of Finite Difference Method

```python

for j in range(len(T)-2, 0, -1):  # time axis(axis = 0)
    for i in range(1, m-1):  # position axis(axis = 1)
        U[j, i] = alpha*U[j+1, i-1] + (1-2*alpha)*U[j+1, i] + alpha*U[j+1, i+1]
```

For plotting the Temperature profile of the Rod at various time stamps

```python
plt.title('Solution of 1D Fourier heat equation when the heatpoint is at x=%.2f' % X[p])
plt.xlabel('Position along the rod')
plt.ylabel(r'Temperature in [$\degree C$]')
s = 0
mark=['o','D','P','^','s','>','<','+',',','.','*']
for k in range(len(T)-1, -1, -20):
    plt.plot(X, U[-k, :], label='t = %d' % T[k],marker = mark[s],markevery = 2)
    s+=1
plt.legend(loc='best')
#plt.savefig('Heat2.png')
plt.show()
```
