```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Creating The grid for the temperature profile at various time at various points

```python
length = 20      #Length of the String
dx = 1/3         # distance between two points on the rod
X = np.arange(0,length,dx)
n = len(X)       #number of positions

time = 100       #duration
dt = 1/36        #smallest time step
T = np.arange(0,time,dt)
m = len(T)          #number of time steps

alpha = 0.25 

U = np.zeros((m,n))     # Creating the grid

BoundaryCondition=[100,0]
InitialCondition = 0
```

Implementation of the initial and boundary conditions

```python
U[0,:] = InitialCondition 
U[:,0] = BoundaryCondition[0]
U[:,-1] = BoundaryCondition[1]
```

Implementation of the finite difference method

```python
for j in range(0,m-1):        #time axis(axis = 0)
    for i in range(1,n-1):  #position axis(axis = 1)
       U[j+1,i] = alpha*(U[j,i+1] + U[j,i-1]) + (1-2*alpha)*U[j,i]
```

```python
mark=['o','D','P','^','s','>','<','+',',','.','*']
plt.title('Solution of 1D Fourier heat equation when the heatpoint is at x=0')
plt.xlabel('Position along the rod')
plt.ylabel(r'Temperature in [$\degree C$]')
for k in range(0, m-1,600):
    plt.plot(X,U[k,:],label = 't = %d'%T[k],marker = mark[int(k/600)],markevery = 3)
plt.legend(loc='best')
#plt.savefig('heat1.png')
plt.show()
```
